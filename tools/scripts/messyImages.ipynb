```python
import os, glob
from scipy.misc import imread
```

```python
def getPoints(path):
    with open(path) as f:
        name = (path.split('.')[0]).split('/')[-1]
        for line in f.readlines():
            image = line.split(' ')[-1]
            
            array = image.split('_')
            
            if(len(array) > 3):
                lat = array[1]
                lon = array[2]
            else:
                
                lat = array[0][2:]
                lon = array[1]
            print lat, lon
            with open(name + '.pointList','a') as f:
                f.write(str((float(lat), float(lon)))+'\n')
        
```
